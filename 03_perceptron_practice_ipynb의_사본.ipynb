```python
import numpy as np
```

```python
# AND 데이터셋 정의
AND_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # 입력값
AND_y = np.array([0, 0, 0, 1])  # AND 출력값

# OR 데이터셋 정의
OR_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # 입력값
OR_y = np.array([0, 1, 1, 1])  # OR 출력값

# XOR 데이터셋 정의
XOR_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # 입력값
XOR_y = np.array([0, 1, 1, 0])  # XOR 출력값
```

```python
# 퍼셉트론 모델 정의
class Perceptron:
    def __init__(self, input_size, lr=0.1, epochs=10):
        self.weights = np.zeros(input_size + 1)  # 가중치 초기화 (bias 포함) > 이해 해야함 <
        self.lr = lr  # 학습률
        self.epochs = epochs  # 학습 반복 횟수

    def activation(self, x):
        return 1 if x >= 0 else 0  # 단위 계단 함수

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]  # 가중치와 입력값 계산
        return self.activation(z)

    def fit(self, X, y): # x 데이터와 y레이블을 받아서 여러번 epoch 돌리면서 퍼셉트론을 학습시킴
        for epoch in range(self.epochs):
            for i in range(X.shape[0]): #  샘플을 순서대로 하나씩 보겠다는 의미
                x_i = X[i]
                y_pred = self.predict(x_i)
                error = y[i] - y_pred

                # 가중치 업데이트
                self.weights[1:] += self.lr * error * x_i
                self.weights[0] += self.lr * error  # bias 업데이트

            # 학습 과정 출력
            print(f"Epoch {epoch + 1}/{self.epochs}, Weights: {self.weights}")
```

# AND 데이터 학습

```python
# 퍼셉트론 학습
perceptron = Perceptron(input_size=2, lr=0.1, epochs=10)
perceptron.fit(AND_X, AND_y) # fit 학습
```

```python
# AND 테스트
for x, target in zip(AND_X, AND_y):
    pred = perceptron.predict(x)
    print(f"Input: {x}, Predicted: {pred}, Target: {target}")
```

# OR 데이터 학습

```python
# 퍼셉트론 학습
perceptron = Perceptron(input_size=2, lr=0.1, epochs=10)
perceptron.fit(OR_X, OR_y)
```

```python
# OR 테스트
for x, target in zip(OR_X, OR_y):
    pred = perceptron.predict(x)
    print(f"Input: {x}, Predicted: {pred}, Target: {target}")
```

# XOR 데이터 학습

```python
# 퍼셉트론 학습
perceptron = Perceptron(input_size=2, lr=0.1, epochs=10)
perceptron.fit(XOR_X, XOR_y)
```

```python
# XOR 문제 테스트
print("Testing XOR Problem:")
for x, target in zip(XOR_X, XOR_y):
    pred = perceptron.predict(x)
    print(f"Input: {x}, Predicted: {pred}, Target: {target}")
```

### 1 선형분리 가능
- AND, OR 가 있고 AND는 둘 다 1일때만 나머지는0, OR은 둘중 하나만 1이어도 1로 둘 다 0이면 0 출력

- 예시. 전구가 1개가 있고 스위치는 2개가 있다.
    
    AND : 스위치가 두개가 on 되어 있을 때 전구가 켜짐
    
    OR  : 스위치 2개중 1개 또는 2개 on 되어 있어도 전구가 켜짐

### 2 선형 분리 불가능
- 1. XOR에서 a=1, b=0 이라 했을 때 a, a는 0이 나오고 a, b는 1 b, b는 0이라 출력 둘중 하나만 1이면 1이 뜨고 둘 다 같으면 0이 출력

- 예시. 전구가 있고 스위치는 2개가 있는데 스위치 1개 on 했을 때 전구는 켜지고 스위치를 2개를 on 했을때 전구가 꺼짐
